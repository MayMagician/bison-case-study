```python
import pandas as pd
df = pd.read_csv("mirrorPublications.csv")
df
```

```python
#voglio fare analisi temporale, quindi estraggo le informazioni necessarie dalla colonna created_date

#inizio convertendo created_date in formato datetime per facilitarmi l'estrazione dei dati
df['created_date'] = pd.to_datetime(df['created_date'])

#aggiungo le colonne
df['year'] = df['created_date'].dt.year
df['month'] = df['created_date'].dt.month
df['day'] = df['created_date'].dt.day
df['weekday'] = df['created_date'].dt.day_name()

df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#raggruppo per anno e mese e conto le occorrenze
monthly_distribution = df.groupby(['year', 'month']).size().reset_index(name='count')
#ordino i risultati
monthly_distribution = monthly_distribution.sort_values(by=['year', 'month'])

print(monthly_distribution)

#creo una colonna combinata per anno e mese in formato "YYYY-MM"
monthly_distribution['year_month'] = monthly_distribution['year'].astype(str) + '-' + monthly_distribution['month'].astype(str).str.zfill(2)

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(data=monthly_distribution, x='year_month', y='count', color='blue')
plt.xticks(rotation=45, ha='right')
plt.title('Distribuzione delle pubblicazioni per mese')
plt.xlabel('Anno e mese')
plt.ylabel('Numero di pubblicazioni')
plt.tight_layout()
plt.show()
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

#inizializzo una lista di stopwords
stop_words = set(stopwords.words('english'))

#funzione per pulire il testo
def clean_text(text):
    #combino e rimuovo newline e tab
    text = text.replace('\n', ' ').replace('\t', ' ')
    #rimuovo caratteri speciali e numeri
    text = re.sub(r'[^\w\s]', '', text)
    #rendo tutto il testo minuscolo
    text = text.lower()
    #tokenizzo il testo
    tokens = word_tokenize(text)
    #rimuovo le stopwords
    tokens = [word for word in tokens if word not in stop_words]
    #riconverto in stringa
    return ' '.join(tokens)

#aggiungo la colonna cleaned_text
df['cleaned_text'] = (df['title'].fillna('') + ' ' + df['body'].fillna('')).apply(clean_text)
df
```

```python
from langdetect import detect, DetectorFactory
from langdetect.lang_detect_exception import LangDetectException

DetectorFactory.seed = 0

#funzione per riconoscere la lingua
def detect_language(text):
    try:
        #rilevo la lingua del testo
        return detect(text)
    except LangDetectException:
        #se non è possibile rilevare la lingua, ritorno la stringa "unknown"
        return 'unknown'

#aggiungo la colonna "language"
df['language'] = df['cleaned_text'].apply(detect_language)
```
