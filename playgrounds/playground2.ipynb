```python
import pandas as pd
df = pd.read_csv("mirrorPublications_featured.csv")
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#raggruppo per anno e mese e conto le occorrenze
monthly_distribution = df.groupby(['year', 'month']).size().reset_index(name='count')
#ordino i risultati
monthly_distribution = monthly_distribution.sort_values(by=['year', 'month'])

#stampo la distribuzione mensile
print(monthly_distribution)

#creo una colonna combinata per anno e mese in formato "YYYY-MM"
monthly_distribution['year_month'] = monthly_distribution['year'].astype(str) + '-' + monthly_distribution['month'].astype(str).str.zfill(2)

#creo e stampo il grafico
plt.figure(figsize=(12, 6))
sns.barplot(data=monthly_distribution, x='year_month', y='count', color='blue')
plt.xticks(rotation=45, ha='right', fontsize=18)
plt.yticks(fontsize=18)
plt.xlabel('Anno e mese', fontsize=18)
plt.ylabel('Numero di pubblicazioni', fontsize=18)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
df = pd.read_csv("mirror_publications_etherscan.csv")
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Creazione della colonna "year_month" nel dataframe principale
df['year_month'] = df['year'].astype(str) + '-' + df['month'].astype(str).str.zfill(2)

# Raggruppo per mese e calcolo la somma della variabile Success
monthly_success = df.groupby('year_month')['Success'].sum().reset_index()

# Creazione del grafico
plt.figure(figsize=(12, 6))
sns.barplot(data=monthly_success, x='year_month', y='Success', color='blue')

plt.xticks(rotation=45, ha='right', fontsize=18)
plt.yticks(fontsize=18)
plt.xlabel('Anno e mese', fontsize=18)
plt.ylabel('Success', fontsize=18)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Mappa inglese -> italiano
giorni_ita = {
    'Monday': 'Lunedì',
    'Tuesday': 'Martedì',
    'Wednesday': 'Mercoledì',
    'Thursday': 'Giovedì',
    'Friday': 'Venerdì',
    'Saturday': 'Sabato',
    'Sunday': 'Domenica'
}

# Conta i giorni della settimana
weekday_counts = df['weekday'].value_counts()

# Ordina i giorni in inglese
ordered_days_eng = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekday_counts = weekday_counts.reindex(ordered_days_eng)

# Rinomina gli indici in italiano
weekday_counts.index = [giorni_ita[day] for day in weekday_counts.index]

# Plot
fig, ax = plt.subplots(figsize=(10, 6))
weekday_counts.plot(kind='bar', ax=ax, color='skyblue', width=0.9)

ax.set_xlabel("Giorno della settimana", fontsize=18)
ax.set_ylabel("Numero di articoli", fontsize=18)
plt.xticks(rotation=45, ha='right', fontsize=18)
plt.yticks(fontsize=18)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Estrarre la colonna dei topic dal DataFrame
topic_columns = [
    'topic_T1: Gaming, Virtual Worlds & Characters',
    'topic_T2: Wallets, Airdrops & Ethereum Tools',
    'topic_T3: Web3, Blockchain & Digital Platforms',
    'topic_T4: DeFi, Market Strategies & Liquidity',
    'topic_T5: Blockchain, Transactions & Smart Contracts',
    'topic_T6: Web3 Launches, Rewards & Creators',
    'topic_T7: Human Thoughts, Emotions & Reflections'
]

# Sommare i valori di ciascun topic
topic_counts = df[topic_columns].sum()

# Plot della distribuzione numerica
fig, ax = plt.subplots(figsize=(10, 6))
topic_counts.plot(kind='bar', ax=ax, color='skyblue')
ax.set_title("Distribuzione dell'appartenenza ai Topic", fontsize=16)
ax.set_xlabel('Topic', fontsize=12)
ax.set_ylabel('Numero di occorrenze', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

#non usato nella tesi
```

<b>Ricreo le matrici e i grafici sinteticamente, semplicemente riportando i risultati</b>

```python
#knn

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Supponiamo che questa sia la tua confusion matrix (ipotizziamo i valori plausibili dai report)
cm = np.array([
    [1263, 254, 144, 64],  # classe 0
    [334, 486, 197, 119],   # classe 1
    [206, 234, 508, 228],   # classe 2
    [125, 150, 284, 779]    # classe 3
])

labels = [0, 1, 2, 3]

plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(cm, annot=True, fmt="d", cmap="Reds", xticklabels=labels, yticklabels=labels, annot_kws={"size": 20})
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('Valori predetti', fontsize=20)
plt.ylabel('Valori reali', fontsize=20)
cbar = heatmap.collections[0].colorbar
cbar.ax.tick_params(labelsize=20)
plt.show()
```
