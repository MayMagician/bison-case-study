```python
import pandas as pd
df = pd.read_csv("mirrorPublications_cleaned.csv")
df
```

```python
import spacy
from textblob import TextBlob #per sentiment analysis
import pandas as pd
from tqdm import tqdm #per barre di caricamento

#inizializzo modello di spacy
nlp = spacy.load("en_core_web_sm")

def process_text(text):
    if isinstance(text, str):  #controllo che il testo sia stringa, perché altrimenti ritorno liste vuote
        doc = nlp(text)
        tokens = [token.text for token in doc]  #tokenizzazione
        lemmas = [token.lemma_ for token in doc]  #lemmatizzazione
        pos_tags = [(token.text, token.pos_) for token in doc]  #POS tagging
        return {
            "tokens": tokens,
            "lemmas": lemmas,
            "pos_tags": pos_tags
            }
    #caso in cui il testo non è una stringa e quindi ritorno liste vuote
    return {
        "tokens": [],
        "lemmas": [],
        "pos_tags": []
    }

#funzione per sentiment analysis
def sentiment_analysis(text):
    if isinstance(text, str): #controllo che il testo sia stringa, altrimenti ritorno None
        blob = TextBlob(text)
        return blob.sentiment.polarity #tra -1 e 1
    return None

tqdm.pandas(desc="Aggiungi colonne...")

#applico la funzione clean
df["processed_cleaned_text"] = df["cleaned_text"].progress_apply(lambda x: process_text(x))

#divido il dizionario restituito in colonne separate
df["cleaned_text_tokenized"] = df["processed_cleaned_text"].apply(lambda x: x["tokens"])
df["cleaned_text_lemmatized"] = df["processed_cleaned_text"].apply(lambda x: x["lemmas"])
df["cleaned_text_POS"] = df["processed_cleaned_text"].apply(lambda x: x["pos_tags"])

#applico la sentiment analysis
df["cleaned_text_sentiment"] = df["cleaned_text"].progress_apply(lambda x: sentiment_analysis(x))
```

```python
df
```

<h1>Aggiungo il numero di parole per titolo e body</h1>

```python
df['words_body'] = df['cleaned_body'].apply(lambda x: len(x.split()) if isinstance(x, str) else 0)
df['words_title'] = df['cleaned_title'].apply(lambda x: len(x.split()) if isinstance(x, str) else 0)
df['words_text'] = df['cleaned_text'].apply(lambda x: len(x.split()) if isinstance(x, str) else 0)
df
```

<h1>Aggiungo una feature testuale: somma normalizzata della TF-IDF delle parole nel post (da cleaned_body), calcolata come la somma della TF-IDF di tutte le parole diviso il numero di parole del post</h1>

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

df['cleaned_body'] = df['cleaned_body'].fillna('')  #sostituisco i NaN con stringhe vuote per non causare errori

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df['cleaned_body'])  #matrice tf-idf su tutto il testo dei vari cleaned_body

tfidf_sum = tfidf_matrix.sum(axis=1).A1  #somma dei valori tf-idf per ogni documento; risultato matrice sparsa, ma A1 la converte in un array denso

df['normalized_tfidf_sum'] = tfidf_sum / df['words_body']
```
