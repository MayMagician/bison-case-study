<h2>Aggiungo informazioni sui valori di Optimism, Tether, USDC, DAI e Arbitrum in USD</h2>

```python
#tutti presi da investing.com
optimism_info = pd.read_csv("optimism.csv")
tether_info = pd.read_csv("tether.csv")
usdc_info = pd.read_csv("usdc.csv")
dai_info = pd.read_csv("dai.csv")
#arbitrum_info = pd.read_csv("arbitrum.csv") ---> scartato perché ha troppe nan

#usdc_info ha 26 NaN su volume, le riempio con la moda, il valore più frequente
mode_value = usdc_info['Vol.'].mode()[0]
usdc_info['Vol.'] = usdc_info['Vol.'].fillna(mode_value)
```

```python
#rinomino le colonne dei df per chiarezza una volta che le unirò al mio df principale. cambio Data in created_date per poterli poi unire al df principale
optimism_info = optimism_info.rename(columns={
    'Data': 'created_date',
    'Ultimo': 'last_optimism_usd',
    'Apertura': 'open_optimism_usd',
    'Massimo': 'max_optimism_usd',
    'Minimo': 'min_optimism_usd',
    'Vol.': 'vol_optimism',
    'Var. %': 'var%_optimism'
})

tether_info = tether_info.rename(columns={
    'Data': 'created_date',
    'Ultimo': 'last_tether_usd',
    'Apertura': 'open_tether_usd',
    'Massimo': 'max_tether_usd',
    'Minimo': 'min_tether_usd',
    'Vol.': 'vol_tether',
    'Var. %': 'var%_tether'
})

usdc_info = usdc_info.rename(columns={
    'Data': 'created_date',
    'Ultimo': 'last_usdc_usd',
    'Apertura': 'open_usdc_usd',
    'Massimo': 'max_usdc_usd',
    'Minimo': 'min_usdc_usd',
    'Vol.': 'vol_usdc',
    'Var. %': 'var%_usdc'
})

dai_info = dai_info.rename(columns={
    'Data': 'created_date',
    'Ultimo': 'last_dai_usd',
    'Apertura': 'open_dai_usd',
    'Massimo': 'max_dai_usd',
    'Minimo': 'min_dai_usd',
    'Vol.': 'vol_dai',
    'Var. %': 'var%_dai'
})

#arbitrum_info = arbitrum_info.rename(columns={
#    'Data': 'created_date',
#    'Ultimo': 'last_arbitrum_usd',
#    'Apertura': 'open_arbitrum_usd',
#    'Massimo': 'max_arbitrum_usd',
#    'Minimo': 'min_arbitrum_usd',
#    'Vol.': 'vol_arbitrum',
#    'Var. %': 'var%_arbitrum'
#})
```

```python
#modifico la colonna "created_date" per cambiare il formato della data da gg.mm.aaaa a aaaa-mm-gg in maniera da essere uguale a created_date del df originale, per poter poi unire facilmente i dataset
optimism_info["created_date"] = pd.to_datetime(optimism_info["created_date"], format="%d.%m.%Y").dt.strftime("%Y-%m-%d")
tether_info["created_date"] = pd.to_datetime(tether_info["created_date"], format="%d.%m.%Y").dt.strftime("%Y-%m-%d")
usdc_info["created_date"] = pd.to_datetime(usdc_info["created_date"], format="%d.%m.%Y").dt.strftime("%Y-%m-%d")
dai_info["created_date"] = pd.to_datetime(dai_info["created_date"], format="%d.%m.%Y").dt.strftime("%Y-%m-%d")
#arbitrum_info["created_date"] = pd.to_datetime(arbitrum_info["created_date"], format="%d.%m.%Y").dt.strftime("%Y-%m-%d")

#unisco i df usando 'created_date' come chiave
df = df.merge(optimism_info[['created_date', 'last_optimism_usd', 'open_optimism_usd', 'max_optimism_usd', 'min_optimism_usd', 'vol_optimism', 'var%_optimism']],
                     on='created_date', how='left')

df = df.merge(tether_info[['created_date', 'last_tether_usd', 'open_tether_usd', 'max_tether_usd', 'min_tether_usd', 'vol_tether', 'var%_tether']],
                            on='created_date', how='left')

df = df.merge(usdc_info[['created_date', 'last_usdc_usd', 'open_usdc_usd', 'max_usdc_usd', 'min_usdc_usd', 'vol_usdc', 'var%_usdc']],
                            on='created_date', how='left')

df = df.merge(dai_info[['created_date', 'last_dai_usd', 'open_dai_usd', 'max_dai_usd', 'min_dai_usd', 'vol_dai', 'var%_dai']],
                            on='created_date', how='left')

#df = df.merge(arbitrum_info[['created_date', 'last_arbitrum_usd', 'open_arbitrum_usd', 'max_arbitrum_usd', 'min_arbitrum_usd', 'vol_arbitrum', 'var%_arbitrum']],
#                            on='created_date', how='left')

#formatto il dataset una volta uniti per essere coerente con le altre colonne e utilizzabili con i miei modelli

#modifico le colonne con i prezzi rimuovendo i punti, sostituendo le virgole con i punti e convertendo in float. 1.856,44 -> 1856.44
to_modify = ['last_optimism_usd', 'open_optimism_usd', 'max_optimism_usd', 'min_optimism_usd',
             'last_tether_usd', 'open_tether_usd', 'max_tether_usd', 'min_tether_usd',
             'last_usdc_usd', 'open_usdc_usd', 'max_usdc_usd', 'min_usdc_usd',
             'last_dai_usd', 'open_dai_usd', 'max_dai_usd', 'min_dai_usd']
#'last_arbitrum_usd', 'open_arbitrum_usd', 'max_arbitrum_usd', 'min_arbitrum_usd'

for col in to_modify:
    df[col] = df[col].str.replace('.', '', regex=False)  #rimuove i punti
    df[col] = df[col].str.replace(',', '.', regex=False)  #sostituisce le virgole con i punti
    df[col] = df[col].astype(float)  #converte in float

#modifico le colonne con i volumi sostituendo le virgole con i punti, rimuovendo K/M e moltiplicando per 1000/1000000. 201,70K -> 2017000.0
to_modify_vol = ['vol_optimism', 'vol_tether', 'vol_usdc', 'vol_dai']
#'vol_arbitrum'
for col in to_modify_vol:
    df[col] = df[col].astype(str).str.replace(',', '.', regex=False)  #sostituisce la virgola con il punto
    df[col] = df[col].apply(lambda x: float(x.replace('K', '').replace('M', '').replace('B', '')) * 1000 if 'K' in x 
                        else (float(x.replace('M', '').replace('K', '').replace('B', '')) * 1000000 if 'M' in x 
                              else (float(x.replace('B', '').replace('M', '').replace('K', '')) * 1000000000 if 'B' in x 
                                    else float(x)))) #converte in float, rimuove K/M/B e moltiplica per 1000, per 1000000 o per 1000000000

#modifico le colonne con la var% sostituendo le virgole con i punti e rimuovendo il simbolo percentuale. -1,24% -> -1.24
to_modify_var = ['var%_optimism', 'var%_tether', 'var%_usdc', 'var%_dai']
#'var%_arbitrum'
for col in to_modify_var:
    df[col] = df[col].str.replace(',', '.', regex=False)  #sostituisce la virgola con il punto
    df[col] = df[col].str.replace('%', '', regex=False)  #rimuove '%'

df
```

<h2>Aggiungo il numero di transazioni giornaliere su Optimism</h2>

```python
#csv ottenuto da https://optimistic.etherscan.io/chart/tx
optimism_transactions = pd.read_csv("optimismTransactions.csv")
optimism_transactions
```

```python
#converto Date(UTC) in datetime
optimism_transactions['Date(UTC)'] = pd.to_datetime(optimism_transactions['Date(UTC)'], format='%m/%d/%Y')

#filtro il df per le date tra il 16/08/2022 e il 18/11/2024
start_date = '2022-08-16'
end_date = '2024-11-18'
optimism_transactions = optimism_transactions[(optimism_transactions['Date(UTC)'] >= start_date) & (optimism_transactions['Date(UTC)'] <= end_date)]

optimism_transactions
```

```python
#evito il warning
optimism_transactions = optimism_transactions.copy()

#rinomino la colonna della data per poter unire il dataset al mio originale
optimism_transactions.rename(columns={'Date(UTC)': 'created_date'}, inplace=True)
#la converto nel tipo stringa perché attualmente è del tipo pandas datetime e non potrei unirla
optimism_transactions['created_date'] = optimism_transactions['created_date'].astype(str)

#droppo la colonna timestamp di cui non ho bisogno
optimism_transactions.drop(columns=['UnixTimeStamp'], inplace=True)

#cambio il nome della colonna Value in un nome più comprensibile quando unirò i df
optimism_transactions.rename(columns={'Value': 'daily_transactions_optimism'}, inplace=True)

optimism_transactions
```

```python
#unisco il df optimism_transactions al mio df per aggiungere il numero di transazioni giornaliere su optimism
df = pd.merge(df, optimism_transactions[['created_date', 'daily_transactions_optimism']], on='created_date', how='left')
df
```

<h2>Aggiungo il punteggio di ricerche su Google (1-100) in quella settimana per NFT, Optimism, Crypto, Bitcoin ed Ethereum</h2>

```python
#csv ottenuti da https://trends.google.it/
google_searches_nft = pd.read_csv("google_searches_nft.csv")
google_searches_crypto = pd.read_csv("google_searches_crypto.csv")
google_searches_bitcoin = pd.read_csv("google_searches_bitcoin.csv")
google_searches_ethereum = pd.read_csv("google_searches_ethereum.csv")
google_searches_optimism = pd.read_csv("google_searches_optimism.csv")
```

```python
google_searches_nft = google_searches_nft.reset_index().rename(columns={'index': 'created_date', 'Categoria: Tutte le categorie': 'week_google_searches_nft'}).drop(index=0).reset_index(drop=True)
google_searches_crypto = google_searches_crypto.reset_index().rename(columns={'index': 'created_date', 'Categoria: Tutte le categorie': 'week_google_searches_crypto'}).drop(index=0).reset_index(drop=True)
google_searches_bitcoin = google_searches_bitcoin.reset_index().rename(columns={'index': 'created_date', 'Categoria: Tutte le categorie': 'week_google_searches_bitcoin'}).drop(index=0).reset_index(drop=True)
google_searches_ethereum = google_searches_ethereum.reset_index().rename(columns={'index': 'created_date', 'Categoria: Tutte le categorie': 'week_google_searches_ethereum'}).drop(index=0).reset_index(drop=True)
google_searches_optimism = google_searches_optimism.reset_index().rename(columns={'index': 'created_date', 'Categoria: Tutte le categorie': 'week_google_searches_optimism'}).drop(index=0).reset_index(drop=True)

google_searches_optimism
```

```python
#converti temporaneamente 'created_date' di df in formato data
df['created_date'] = pd.to_datetime(df['created_date'])

#crea una colonna 'week_start_date' nei miei dataset che rappresenta l'inizio della settimana
google_searches_nft['created_date'] = pd.to_datetime(google_searches_nft['created_date'])
google_searches_nft['week_start_date'] = google_searches_nft['created_date'] - pd.to_timedelta(google_searches_nft['created_date'].dt.weekday, unit='D')
google_searches_crypto['created_date'] = pd.to_datetime(google_searches_crypto['created_date'])
google_searches_crypto['week_start_date'] = google_searches_crypto['created_date'] - pd.to_timedelta(google_searches_crypto['created_date'].dt.weekday, unit='D')
google_searches_bitcoin['created_date'] = pd.to_datetime(google_searches_bitcoin['created_date'])
google_searches_bitcoin['week_start_date'] = google_searches_bitcoin['created_date'] - pd.to_timedelta(google_searches_bitcoin['created_date'].dt.weekday, unit='D')
google_searches_ethereum['created_date'] = pd.to_datetime(google_searches_ethereum['created_date'])
google_searches_ethereum['week_start_date'] = google_searches_ethereum['created_date'] - pd.to_timedelta(google_searches_ethereum['created_date'].dt.weekday, unit='D')
google_searches_optimism['created_date'] = pd.to_datetime(google_searches_optimism['created_date'])
google_searches_optimism['week_start_date'] = google_searches_optimism['created_date'] - pd.to_timedelta(google_searches_optimism['created_date'].dt.weekday, unit='D')

#esegui merge_asof per unire il primo df sulla base della settimana
df = pd.merge_asof(df.sort_values('created_date'), 
                          google_searches_nft.sort_values('created_date'),
                          on='created_date',
                          direction='backward')

#rimuovi week_start_date per poter unire anche gli altri dataset senza problemi
df = df.drop(columns=['week_start_date'])

#esegui merge_asof per unire il secondo df sulla base della settimana
df = pd.merge_asof(df.sort_values('created_date'), 
                          google_searches_crypto.sort_values('created_date'),
                          on='created_date',
                          direction='backward')

#rimuovi week_start_date per poter unire anche gli altri dataset senza problemi
df = df.drop(columns=['week_start_date'])

#esegui merge_asof per unire il terzo df sulla base della settimana
df = pd.merge_asof(df.sort_values('created_date'), 
                          google_searches_bitcoin.sort_values('created_date'),
                          on='created_date',
                          direction='backward')

#rimuovi week_start_date per poter unire anche gli altri dataset senza problemi
df = df.drop(columns=['week_start_date'])

#esegui merge_asof per unire il quarto df sulla base della settimana
df = pd.merge_asof(df.sort_values('created_date'), 
                          google_searches_ethereum.sort_values('created_date'),
                          on='created_date',
                          direction='backward')

#rimuovi week_start_date per poter unire anche gli altri dataset senza problemi
df = df.drop(columns=['week_start_date'])

#esegui merge_asof per unire il quinto df sulla base della settimana
df = pd.merge_asof(df.sort_values('created_date'), 
                          google_searches_optimism.sort_values('created_date'),
                          on='created_date',
                          direction='backward')

#rimuovi week_start_date definitivamente perché non mi serve più
df = df.drop(columns=['week_start_date'])

#riconverti created_date in str
df['created_date'] = df['created_date'].astype(str)

df
```

```python
import numpy as np

df.drop(columns=['Success'], inplace=True)
df.drop(columns=['SuccessBinary'], inplace=True)

df.loc[:, 'Success'] = df['Total Sold Numbers'] * df['Price(ETH)'] * 100000

#threshold basata sulla mediana di success
threshold = df['Success'].median()

#creo la colonna SuccessBinary per classificazione
df['SuccessBinary'] = np.where(df['Success'] >= threshold, 1, 0)

counts = df['SuccessBinary'].value_counts()
print(counts)

df
```
