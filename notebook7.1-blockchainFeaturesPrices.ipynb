Come nuove features prendo:

- Informazioni su Bitcoin nel giorno di pubblicazione

- Informazioni su Ether nel giorno di pubblicazione

Per farlo ho preso dei .csv sui prezzi di Bitcoin da investing.it (https://it.investing.com/crypto/ethereum/historical-data per Ether e https://it.investing.com/crypto/bitcoin/historical-data per Bitcoin), che sono così composte:

- Data ->	18.11.2024

- Ultimo -> 3.207,46

- Apertura -> 3.075,42

- Massimo -> 3.220,65

- Minimo -> 3.052,29

- Vol. -> 698,67K (volume di scambi del giorno, cioè la quantità di BTC/ETH che è stata compravenduta in quel giorno)

- Var. % -> 4,29% (variazione percentuale del prezzo rispetto al giorno precedente)

```python
btc_info = pd.read_csv("btc.csv")
eth_info = pd.read_csv("eth.csv")
```

```python
btc_info
```

```python
eth_info
```

```python
#rinomino le colonne dei df per chiarezza una volta che le unirò al mio df principale. cambio Data in created_date per poterli poi unire al df principale
btc_info = btc_info.rename(columns={
    'Data': 'created_date',
    'Ultimo': 'last_btc_usd',
    'Apertura': 'open_btc_usd',
    'Massimo': 'max_btc_usd',
    'Minimo': 'min_btc_usd',
    'Vol.': 'vol_btc',
    'Var. %': 'var%_btc'
})

eth_info = eth_info.rename(columns={
    'Data': 'created_date',
    'Ultimo': 'last_eth_usd',
    'Apertura': 'open_eth_usd',
    'Massimo': 'max_eth_usd',
    'Minimo': 'min_eth_usd',
    'Vol.': 'vol_eth',
    'Var. %': 'var%_eth'
})

#modifico la colonna "created_date" per cambiare il formato della data da gg.mm.aaaa a aaaa-mm-gg in maniera da essere uguale a created_date del df originale, per poter poi unire facilmente i dataset
btc_info["created_date"] = pd.to_datetime(btc_info["created_date"], format="%d.%m.%Y").dt.strftime("%Y-%m-%d")
eth_info["created_date"] = pd.to_datetime(eth_info["created_date"], format="%d.%m.%Y").dt.strftime("%Y-%m-%d")
```

```python
btc_info
```

```python
eth_info
```

```python
# Unisci i DataFrame usando 'created_date' come chiave
df = df.merge(btc_info[['created_date', 'last_btc_usd', 'open_btc_usd', 'max_btc_usd', 'min_btc_usd', 'vol_btc', 'var%_btc']],
                     on='created_date', how='left')

df = df.merge(eth_info[['created_date', 'last_eth_usd', 'open_eth_usd', 'max_eth_usd', 'min_eth_usd', 'vol_eth', 'var%_eth']],
                            on='created_date', how='left')

df
```

```python
df.to_csv('mirrorPublications_prices.csv', index=False)
```
