```python
# Controlla se ci sono NaN e dove
nan_info = df.isna().sum()
print("NaN values per column:\n", nan_info)
nan_info
```

```python
#filtro le righe per cui author_transactions_number è NaN
nan_transactions = df[df["author_transactions_number"].isna()]

#controllo se in tutte queste righe author_ether_balance è 0
all_zero_balance = (nan_transactions["author_ether_balance"] == 0.0).all()

if all_zero_balance:
    print("In tutti i casi in cui 'author_transactions_number' è NaN, 'author_ether_balance' è 0.")
else:
    print("Esistono casi in cui 'author_transactions_number' è NaN, ma 'author_ether_balance' non è 0.")
```

```python
unexpected_cases = df[(df["author_transactions_number"].isna()) & (df["author_ether_balance"] != 0.0)]
unexpected_cases
```

```python
#se 'author_transactions_number' è NaN, 'author_ether_balance' è 0, tranne che per 11 casi, che sono abbastanza pochi da poter essere ignorati
#sostituisco quindi tutti i NaN in author_transactions_number con 0
df["author_transactions_number"] = df["author_transactions_number"].fillna(0.0)
```
