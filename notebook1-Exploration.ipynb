```python
import pandas as pd
df = pd.read_csv("mirrorPublications_withFeatures.csv")
df
```

<h1>Riepilogo statistico iniziale del dataset</h1>

```python
#creo un df_filtered per escludere le colonne che non ha senso considerare nel riepilogo statistico
filtered_df = df.drop(columns=['timestamp', 'month', 'year', 'day'], errors='ignore')

#applico describe sul df_filtered
print(filtered_df.describe())
```

<h1>Analisi di distribuzione linguistica: presenza delle varie lingue nei post</h1>

```python
import pandas as pd
import matplotlib.pyplot as plt

#in language_counts metto la frequenza delle lingue
language_counts = df['language'].value_counts()

#visualizzo le prime dieci
top_languages = language_counts.head(10)
print(top_languages)

plt.figure(figsize=(10, 6))
top_languages.plot(kind='bar', color='skyblue')
plt.title('Lingue più comuni nel dataset')
plt.xlabel('Lingua')
plt.ylabel('Numero di occorrenze')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#rimuovo tutte le righe in cui la lingua non è l'inglese, per essere sicuro di avere solamente dati significativi, confrontabili e interpretabili d'ora in poi
df = df[df["language"] == "en"]
df["language"].value_counts()
```

<h1>Analisi di distribuzione temporale: occorrenza dei post nei vari mesi</h1>

```python
import matplotlib.pyplot as plt
import seaborn as sns

#raggruppo per anno e mese e conto le occorrenze
monthly_distribution = df.groupby(['year', 'month']).size().reset_index(name='count')
#ordino i risultati
monthly_distribution = monthly_distribution.sort_values(by=['year', 'month'])

#stampo la distribuzione mensile
print(monthly_distribution)

#creo una colonna combinata per anno e mese in formato "YYYY-MM"
monthly_distribution['year_month'] = monthly_distribution['year'].astype(str) + '-' + monthly_distribution['month'].astype(str).str.zfill(2)

#creo e stampo il grafico
plt.figure(figsize=(12, 6))
sns.barplot(data=monthly_distribution, x='year_month', y='count', color='blue')
plt.xticks(rotation=45, ha='right')
plt.title('Distribuzione delle pubblicazioni per mese')
plt.xlabel('Anno e mese')
plt.ylabel('Numero di pubblicazioni')
plt.tight_layout()
plt.show()
```

<h1>Analisi di correlazione: stampa di una Correlation Matrix per vedere eventuali correlazioni tra variabili numeriche</h1>

```python
import seaborn as sns
import matplotlib.pyplot as plt

#creo un numerical_df contenente solo le colonne numeriche
numerical_df = df.select_dtypes(include=['float64', 'int64'])

#creo la correlation matrix
correlation_matrix = numerical_df.corr()

#visualizzo la heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix tra variabili numeriche")
plt.show()
```

A parte le ovvie correlazioni tra timestamp e i suoi derivati, le uniche correlazioni degne di nota risultano essere:
- Price(ETH) e Total Sold(ETH) con 0.29.
- Total Sold Numbers e Total Buyers con 0.23, che mi sarei aspettato essere più grande.

<h1>Analisi frequenza delle parole: ricerca parole più comuni e wordcloud</h1>

```python
from collections import Counter

#creo una stringa unica con le parole di tutti i post, sostituendo gli eventuali NaN con stringhe vuote e convertengo tutto sempre in stringa
text = ' '.join(df['cleaned_text'].fillna('').astype(str))
#creo una lista di parole con text.split() e applico Counter per contare la frequenza di ogni elemento della lista (quindi di ogni parola)
word_counts = Counter(text.split())
#stampo le 10 parole più comuni
print(word_counts.most_common(10))
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

#creo un wordcloud sulla base della stringa unica prima generata
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
