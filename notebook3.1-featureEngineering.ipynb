```python
import pandas as pd
df = pd.read_csv("mirrorPublications_withTopics.csv")
df
```

<h3>Voglio aggiungere delle features legate alle parole più importanti nel dataset. Comincio con l'estrarre le 10 parole di tutto il documento con punteggio TF-IDF più alto.</h3>

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

#estraggo il testo da cleaned_text
documents = df['cleaned_text'].astype(str).tolist()
#calcolo la matrice tf-idf
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)
#ottengo le parole
feature_names = vectorizer.get_feature_names_out()
#identifico il massimo TF-IDF score per ogni parola, sommando i punteggi su tutti i documenti per ottenere un ranking globale
tfidf_sums = tfidf_matrix.sum(axis=0).A1  #sommo lungo l'asse dei documenti
word_scores = list(zip(feature_names, tfidf_sums))

#stampo le top10
top_words = sorted(word_scores, key=lambda x: x[1], reverse=True)[:10]
for word, score in top_words:
    print(f"Parola: {word}, TF-IDF: {score}")
```

```python
#aggiungo le parole più significative della mia top10 alla lista di keywords
keywords = ['nft', 'web3', 'community', 'blockchain', 'crypto', 'wallet', 'chain']
#per ogni keyword, creo una nuova colonna, il cui valore sarà 1 se la parola è presente nel testo e 0 nel caso contrario
for keyword in keywords:
    df[keyword] = df['cleaned_text'].apply(lambda text: 1 if keyword in text else 0)
df
```

```python
#guardo i risultati delle keywords
value_counts = {keyword: df[keyword].sum() for keyword in keywords}
value_counts
```

<h3>Aggiungo 3 features per vedere la densità di alcune POS (verb, adjective, noun) nel mio dataframe</h3>

```python
#trasformo cleaned_text_POS in una lista di tuple (attualmente è una stringa) attraverso la libreria ast
import ast

#funzione per convertire una stringa in una lista di tuple
def convert_to_tuples(text):
    try:
        return ast.literal_eval(text)
    except:
        return [] #lista vuota in caso di errore

#applico la funzione a ogni elemento del df
df['cleaned_text_POS'] = df['cleaned_text_POS'].apply(convert_to_tuples)
df
```

```python
def calculate_density(pos_list, pos_target):
    #numero totale di parole
    total_tuples = len(pos_list)
    #numero totale di parole con POS corrispondente al target
    verb_count = sum(1 for word, pos in pos_list if pos == pos_target)
    #ritorno la densità
    return verb_count / total_tuples if total_tuples > 0 else 0

#aggiungo le nuove features
df['verbs_density'] = df['cleaned_text_POS'].apply(lambda x: calculate_density(x, "VERB"))
df['adjectives_density'] = df['cleaned_text_POS'].apply(lambda x: calculate_density(x, "ADJ"))
df['nouns_density'] = df['cleaned_text_POS'].apply(lambda x: calculate_density(x, "NOUN"))
```

<h3>Aggiungo come feature l'indice Flesch-Kincaid, ovvero un indice di leggibilità del testo, che può essere correlato alla sua tecnicità (più è alto, più il testo è semplice da leggere)</h3>

```python
from textstat import textstat

#funziomne per estrarre l'indice flesch-kincaid del testo
def readability_index(text):
    return textstat.flesch_kincaid_grade(text)

df['readability_index'] = df['body'].apply(readability_index) #applico la funzione sul body senza la pulizia

df
```

```python
count_readability_index = df['readability_index'].value_counts()
count_readability_index
```

```python
min_value = df['readability_index'].min()
print(min_value)
max_value = df['readability_index'].max()
print(max_value)
```
