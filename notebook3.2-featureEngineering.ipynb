```python
import pandas as pd
df = pd.read_csv("mirrorPublications_featured_new.csv")
df
```

<h3>Utilizzo la libreria readability per estrarre ulteriori features. Dal momento che tra queste c'è anche l'indice Flesch-Kincaid, droppo l'attuale colonna readability_index e la vado a re-implementare con la nuova libreria, anche per coerenza rispetto alle altre features che estraggo.</h3>

```python
df = df.drop(columns=["readability_index"])
```

```python
df.iloc[0]
```

```python
from readability import getmeasures
def calculate_readability(text):
    try:
        measures = getmeasures(text, lang='en')
        readability_grades = measures.get('readability grades', {})
        sentence_info = measures.get('sentence info', {})

        #ritorna un dizionario con tutte le caratteristiche elencate
        return {
            'Kincaid': readability_grades.get('Kincaid'),
            'FleschReadingEase': readability_grades.get('FleschReadingEase'),
            'GunningFogIndex': readability_grades.get('GunningFogIndex'),
            'characters_per_word': sentence_info.get('characters_per_word'),
            'syll_per_word': sentence_info.get('syll_per_word'),
            'words_per_sentence': sentence_info.get('words_per_sentence'),
            'sentences_per_paragraph': sentence_info.get('sentences_per_paragraph'),
            'type_token_ratio': sentence_info.get('type_token_ratio'),
            'characters': sentence_info.get('characters'),
            'syllables': sentence_info.get('syllables'),
            'words': sentence_info.get('words'),
            'wordtypes': sentence_info.get('wordtypes'),
            'sentences': sentence_info.get('sentences'),
            'paragraphs': sentence_info.get('paragraphs'),
            'long_words': sentence_info.get('long_words'),
            'complex_words': sentence_info.get('complex_words')
        }
    except Exception as e:
        #ritorno None in caso di testo non valido
        print(f"Errore nel calcolo della leggibilità: {e}")
        return {key: None for key in [
            'Kincaid', 'FleschReadingEase', 'GunningFogIndex', 'characters_per_word',
            'syll_per_word', 'words_per_sentence', 'sentences_per_paragraph',
            'type_token_ratio', 'characters', 'syllables', 'words', 'wordtypes',
            'sentences', 'paragraphs', 'long_words', 'complex_words'
        ]}

readability_features = df['body'].apply(calculate_readability)
readability_features
```

```python
#converto readability_features in un dataframe e lo aggiungo al df originale
readability_df = pd.DataFrame(list(readability_features))
df = pd.concat([df, readability_df], axis=1)
df
```

```python
columns_to_check = [
    'Kincaid', 'FleschReadingEase', 'GunningFogIndex', 'characters_per_word',
    'syll_per_word', 'words_per_sentence', 'sentences_per_paragraph',
    'type_token_ratio', 'characters', 'syllables', 'words', 'wordtypes',
    'sentences', 'paragraphs', 'long_words', 'complex_words'
]
none_counts = df[columns_to_check].isna().sum()
none_counts
```

```python
#controllo che tipo di righe sono quelle in cui le nuove features sono NaN
df_copy = df[df[columns_to_check].isna().any(axis=1)]
df_copy
```

```python
#effettivamente le nuove features sono NaN nelle righe in cui il body non contiene parole, come previsto. Sostituisco i NaN con 0.

df[columns_to_check] = df[columns_to_check].fillna(0)

df_copy = df[df[columns_to_check].isna().any(axis=1)]
df_copy
```
