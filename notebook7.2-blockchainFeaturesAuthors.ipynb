```python
import pandas as pd
import numpy as np

df = pd.read_csv("mirrorPublications_prices.csv")

df
```

```python
# Conta le occorrenze degli autori più frequenti
author_counts = df['Author Homepage'].value_counts()

# Mostra i primi 10 autori più frequenti
print("Top 10 autori più frequenti:")
print(author_counts.head(10))

# Analizza il numero totale di NFT venduti per autore
author_sales = df.groupby('Author Homepage')['Total Sold Numbers'].sum().sort_values(ascending=False)

# Mostra i primi 10 autori con più vendite
print("\nTop 10 autori con più NFT venduti:")
print(author_sales.head(10))

# Analizza il totale di ETH guadagnati per autore
author_revenue = df.groupby('Author Homepage')['Total Sold(ETH)'].sum().sort_values(ascending=False)

# Mostra i primi 10 autori con il maggior guadagno
print("\nTop 10 autori con più guadagni in ETH:")
print(author_revenue.head(10))

# Conta il numero di autori unici
unique_authors = df['Author Homepage'].nunique()
print("\nNumero totale di autori singoli:", unique_authors)
```

```python
# Conta il numero di post per autore
author_counts = df['Author Homepage'].value_counts()

# Filtra solo gli autori con più di un post
multiple_posts_authors = author_counts[author_counts > 1]

# Mostra il numero di autori con più di un post
print(f"Numero di autori con più di un post: {len(multiple_posts_authors)}")

# Mostra i primi 10 autori con più post
print(multiple_posts_authors.head(10))
```

```python
#aggiungo come feature il numero di post dell'autore nel df
df['authorPostCount'] = df['Author Homepage'].map(df['Author Homepage'].value_counts())

num_zero_count = (df['authorPostCount'] == 0).sum()
print(f"Numero di volte in cui 'authorPostCount' è 0: {num_zero_count}")

num_one_count = (df['authorPostCount'] == 1).sum()
print(f"Numero di volte in cui 'authorPostCount' è 1: {num_one_count}")

df
```

```python
#aggiungo come feature il numero di wNFT venduti (quindi quante volte un suo post è stato mintato nel dataset)

#numero totale di NFT venduti per autore
author_sales_count = df.groupby('Author Homepage')['Total Sold Numbers'].sum()

df['authorTotalSales'] = df['Author Homepage'].map(author_sales_count)

num_zero_sales = (df['authorTotalSales'] == 0).sum()
print(f"Numero di volte in cui 'authorTotalSales' è 0: {num_zero_sales}")

num_one_sales = (df['authorTotalSales'] == 1).sum()
print(f"Numero di volte in cui 'authorTotalSales' è 1: {num_one_sales}")

df
```

```python
#aggiungo come feature il numero di guadagni in ETH (quindi quanti ETH ha guadagnato per le sue vendite nel dataset)

author_revenue = df.groupby('Author Homepage')['Total Sold(ETH)'].sum()

df['authorTotalRevenue'] = df['Author Homepage'].map(author_revenue)

num_zero_revenue = (df['authorTotalRevenue'] == 0).sum()
print(f"Numero di volte in cui 'Author Total Revenue' è 0: {num_zero_revenue}")

df
```

<h3>Prima di salvare il df, mi sono accorto che la formattazione dei numeri nelle colonne nuove sui prezzi di BTC ed ETH è sbagliata, in quanto usa i punti per chiarezza ogni 3 numeri e le virgole invece dei punti come separatore decimale + altre cose specifiche per ogni colonna.

Modifico il formato in modo da non avere problemi in futuro prima di salvare il dataset.</h3>

```python
#modifico le colonne con i prezzi rimuovendo i punti, sostituendo le virgole con i punti e convertendo in float. 1.856,44 -> 1856.44
to_modify = ['last_btc_usd', 'open_btc_usd', 'max_btc_usd', 'min_btc_usd', 
                   'last_eth_usd', 'open_eth_usd', 'max_eth_usd', 'min_eth_usd']

for col in to_modify:
    df[col] = df[col].str.replace('.', '', regex=False)  #rimuove i punti
    df[col] = df[col].str.replace(',', '.', regex=False)  #sostituisce le virgole con i punti
    df[col] = df[col].astype(float)  #converte in float

#modifico le colonne con i volumi sostituendo le virgole con i punti, rimuovendo K/M e moltiplicando per 1000/1000000. 201,70K -> 2017000.0
to_modify_vol = ['vol_btc', 'vol_eth']

for col in to_modify_vol:
    df[col] = df[col].str.replace(',', '.', regex=False)  #sostituisce la virgola con il punto
    df[col] = df[col].apply(lambda x: float(x.replace('K', '').replace('M', '')) * 1000 if 'K' in x else (float(x.replace('M', '').replace('K', '')) * 1000000 if 'M' in x else float(x))) #converte in float, rimuove K/M e moltiplica per 1000 o per 1000000

#modifico le colonne con la var% sostituendo le virgole con i punti e rimuovendo il simbolo percentuale. -1,24% -> -1.24
to_modify_var = ['var%_btc', 'var%_eth']
for col in to_modify_var:
    df[col] = df[col].str.replace(',', '.', regex=False)  #sostituisce la virgola con il punto
    df[col] = df[col].str.replace('%', '', regex=False)  #rimuove '%'
```
